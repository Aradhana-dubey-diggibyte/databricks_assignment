```python
from pyspark.sql.functions import col
```

```python
silver_dim_config = {
    "dim_customer": {
        "bronze_table": "learning_lab_catalog.bronze.dim_customer",
        "silver_table": "learning_lab_catalog.silver.dim_customer",
        "business_key": "Customer_ID"
    },
    "dim_branch": {
        "bronze_table": "learning_lab_catalog.bronze.dim_branch",
        "silver_table": "learning_lab_catalog.silver.dim_branch",
        "business_key": "Branch_ID"
    },
    "dim_transaction_type": {
        "bronze_table": "learning_lab_catalog.bronze.dim_transaction_type",
        "silver_table": "learning_lab_catalog.silver.dim_transaction_type",
        "business_key": "Type_ID"
    },
    "dim_transaction_status": {
        "bronze_table": "learning_lab_catalog.bronze.dim_transaction_status",
        "silver_table": "learning_lab_catalog.silver.dim_transaction_status",
        "business_key": "Status_ID"
    },
    "dim_date": {
        "bronze_table": "learning_lab_catalog.bronze.dim_date",
        "silver_table": "learning_lab_catalog.silver.dim_date",
        "business_key": "Date_ID"
    }
}
```

```python
for dim_name, cfg in silver_dim_config.items():

    df_bronze = spark.table(cfg["bronze_table"])

    df_silver = (
        df_bronze
        .dropDuplicates([cfg["business_key"]])
        .filter(col(cfg["business_key"]).isNotNull())
    )

    df_silver.write \
        .format("delta") \
        .mode("overwrite") \
        .option("overwriteSchema", "true") \
        .saveAsTable(cfg["silver_table"])

    print(f"Silver load completed for {dim_name}")
```
